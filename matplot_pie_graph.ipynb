```python
import matplotlib.pyplot as plt
```

```python
sizes = [3,4,6,2]
plt.pie(sizes)
plt.show()
```

```python
sizes = [3,4,6,2]
plt.pie(sizes)
plt.axis("equal")
plt.show()
```

```python
sizes = [3,4,6,2]
# adding own colors
color = ["blue", "red","purple","pink"]
plt.pie(sizes, colors=color)
plt.show()
```

```python
# labels
labels = ["A","B","C","D"]
sizes = [3,4,6,2]
color = ["blue", "red","purple","pink"]
plt.pie(sizes, colors=color, labels = labels)
plt.show()
```

```python
labels = ["A","B","C","D"]
sizes = [3,4,6,2]
color = ["blue", "red","purple","pink"]
# title
plt.title("Split among classes")
plt.pie(sizes, colors=color, labels = labels)
plt.show()
```

```python
labels = ["A","B","C","D"]
sizes = [3,4,6,2]
color = ["blue", "red","purple","pink"]
# title
plt.title("Split among classes")
# addint text on each charts use autopct
# here %.2f means after decimal how many digits we want to show
plt.pie(sizes, colors=color, labels = labels, autopct="%.2f")
plt.show()
```

```python
labels = ["A","B","C","D"]
sizes = [3,4,6,2]
color = ["blue", "red","purple","pink"]
# title
plt.title("Split among classes")
# to add % symbol/sign
# NOTE in autopct we can pass a function also to decide each cell

plt.pie(sizes, colors=color, labels = labels, autopct="%.2f%%")
plt.show()
```

```python
labels = ["A","B","C","D"]
sizes = [3,4,6,2]
color = ["blue", "red","purple","pink"]
# explode function
explode = [0.1,0,0,0]

plt.title("Split among classes")

plt.pie(sizes, colors=color, explode = explode,labels = labels, autopct="%.2f%%")
plt.show()
```

```python
labels = ["A","B","C","D"]
sizes = [3,4,6,2]
color = ["blue", "red","purple","pink"]
# explode function
explode = [0,0,0.2,0]

plt.title("Split among classes")

plt.pie(sizes, colors=color, explode = explode,labels = labels, autopct="%.2f%%")
plt.show()
```

```python
# by default the direction is anti clock wise
#to make it clock wise
#use counterclock = False (by default it is true)

labels = ["A","B","C","D"]
sizes = [3,4,6,2]
color = ["blue", "red","purple","pink"]
# explode function
explode = [0,0,0.2,0]

plt.title("Split among classes")

plt.pie(sizes, colors=color, explode = explode,labels = labels, autopct="%.2f%%", counterclock=False)
plt.show()
```

```python
# the first entry A is starting from 0 we can change that too
# using startangle

labels = ["A","B","C","D"]
sizes = [3,4,6,2]
color = ["blue", "red","purple","pink"]
# explode function
explode = [0,0,0.2,0]

plt.title("Split among classes")

plt.pie(sizes, colors=color, explode = explode,labels = labels, autopct="%.2f%%", counterclock=False, startangle=100)
plt.show()
```

```python
# the first entry A is starting from 0 we can change that too
# using startangle

labels = ["A","B","C","D"]
sizes = [3,4,6,2]
color = ["blue", "red","purple","pink"]
# explode function
explode = [0,0,0.2,0]

plt.title("Split among classes")
# shadow property
plt.pie(sizes, colors=color,shadow=True ,explode = explode,labels = labels, autopct="%.2f%%", counterclock=False, startangle=100)
plt.show()
```
