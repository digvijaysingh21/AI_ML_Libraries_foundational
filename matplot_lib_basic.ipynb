```python
# plot x ^ 3

import numpy as np

x = np.array([1,2,3,4])
y = x **3

plt.scatter(x,y)

plt.plot(x,y)
```

```python
# plot x ^ 3

import numpy as np

# 1st is starting point
# 2nd is ending point
# 3rd is increment
x = np.arange(0,5,0.1)
y = x **3

plt.scatter(x,y)
plt.plot(x,y)
```
