```python
import matplotlib.pyplot as plt
```

```python
import numpy as np

x = np.arange(0,5,0.1)
y = x **3
# marker possible with alphabets like (o,p,s,d,h,x,v)
# markers possibel with symbols like (<,>,_,+,^,*)
# markeres possible with numbers like (1,2,3,4,8)
plt.plot(x,y, color = 'black', marker="o")
```

```python
import numpy as np

x = np.arange(0,5,0.1)
y = x **3
#linewidth
plt.plot(x,y, color = 'black', linewidth = 5)
```

```python
x = np.arange(0,5,0.1)
y = x **3
plt.plot(x,y, color = 'black', linewidth = 5)

# y label

plt.ylabel ("x^3")

# x label
plt.xlabel("x")

# title
plt.title("Matplot learning ")


# plt.show()

```

```python
x = np.arange(0,5,0.1)
y = x **3
plt.plot(x,y, color = 'black', linewidth = 2, label="x^3")

y2 = x**2

plt.plot(x,y2, color="red", linewidth=1.5, label="x^2")

# y label

plt.ylabel ("y")

# x label
plt.xlabel("x")

# title
plt.title("Matplot learning ")
# use legend and label for naming each line
plt.legend()
```

```python
x = np.arange(0,5,0.1)
y = x **3
plt.plot(x,y, color = 'black', linewidth = 2, label="x^3")

y2 = x**2

plt.plot(x,y2, color="red", linewidth=1.5, label="x^2")

# use axis to define own parameter for y and x axis
# it takes 4 paramters as list
#it take 4 parameter basicallt the start and end point for both y and x axis
plt.axis([0,10,0,200])

# y label

plt.ylabel ("y")

# x label
plt.xlabel("x")

# title
plt.title("Matplot learning ")
# use legend and label for naming each line
plt.legend()
```
