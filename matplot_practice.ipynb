**Scatter**

```python
import numpy as np
x = np.array([1,3,4,5,7])
y = np.array([4,1,9,7,2])

plt.scatter(x,y)
plt.plot(x,y, 'ro')
```

**Subplots**

```python
import numpy as np
import matplotlib.pyplot as plt

# plot 1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(2,1,1)
plt.plot(x,y)

# plot 2

x1 = np.array([0,1,2,3,4])
y1 = np.array([10,20,30,40,50])

plt.subplot(2,1,2)
plt.plot(x1,y1)


```
